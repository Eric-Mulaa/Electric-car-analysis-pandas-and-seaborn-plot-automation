#### Automating Plots using python function  

```python
def hbox(data, column):
    '''This function will take two arguments and make histplot and boxplot'''
    # First close any figures that might have been open
    plt.close('all')
    plt.figure(figsize=(10, 6))
    # Defining the subplot with 1 row, 2 columns and plotting a histplot on the subplot index 1
    plt.subplot(1, 2, 1)
    sns.histplot(data = data, x = column, bins=10, kde=True, color = '#911116', line_kws={'color': '#911156'})
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {column}')
    # Plotting a box plot on the second index grind
    plt.subplot(1, 2, 2)
    sns.boxplot(data = data, y = column)
    plt.ylabel(column)
    plt.title(f' Box Plot of {column}')
    # Ensuring that there will be no overlapping within my plot
    plt.tight_layout()
    plt.show()
```

```python
# Ignoring all user category warnings that appear in outputs, these are not errors but warnings
warnings.filterwarnings('ignore', category=UserWarning)
hbox(df, 'Model Year')
hbox(df, 'Electric Range')
hbox(df, 'Base MSRP')
```

We will now see the trend of electric vehicles release over the years.

```python
plt.figure(figsize=(10,6))
df['Model Year'].value_counts().sort_index().plot(kind='line')
plt.ylabel('COUNT')
plt.title('Production Trend')
plt.show()
```

Lets see how county have been adopting electric vehicles over time

```python
#This code snippet calculates the count of each unique value in the 'County' column using value_counts(), 
# sorts the resulting counts in descending order using sort_values(ascending=False), 
# and retrieves the index to get the unique values in the desired descending order.
plt.figure(figsize=(10, 6))
order = df['County'].value_counts().sort_values(ascending=False).index
sns.catplot(data=df, x = 'County', kind = 'count', order=order, height = 6, aspect = 2)
plt.ylabel('COUNT')
plt.title('County Adaption of Electric Vehicles')
plt.xticks(rotation=90)
plt.show
```

The following code will visualize popular makes, top 5.

```python
popular_make = df["Make"].value_counts().head(5)
popular_make.plot(kind='bar')
plt.ylabel('count')
plt.title('Popular Make')
```
